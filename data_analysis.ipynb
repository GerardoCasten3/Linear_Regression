This is a very simple linear regression model based on Housing Price in the United States.

```python
#Import libraries and modules
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

#Load the data
df = pd.read_csv('price.csv')
df
```

```python
#Set variables
X = df['SquareFeet'].values.reshape(-1,1)
Y = df['Price'].values.reshape(-1,1)

#Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X, Y)
# Line equation
intercept = lin_reg.intercept_
slope = lin_reg.coef_
print('Line equation: Y =', round(slope[0][0],5), 'x +', round(intercept[0],5))
print('Correlation Coefficient:', round(np.corrcoef(df['SquareFeet'], df['Price'])[0,1],5))
print('Determination Coefficient:', round(r2_score(df['Price'], lin_reg.predict(df[['SquareFeet']])), 5))


#Create a scatter plot
sns.regplot(x=X, y=Y, ci=95, line_kws={'color':'red'}, scatter_kws={'color':'black'})  
plt.title("Linear Regression Model with 95% Confidence Interval")
plt.xlabel('Square Feet')
plt.ylabel('Price in USD')
plt.show()

#Analyze if the neighborhood has an impact on the price
sns.scatterplot(x=df['SquareFeet'], y=df['Price'], hue=df['Neighborhood'])
plt.title("Relation between Square Feet and Price by Neighborhood")
plt.xlabel('Square Feet')
plt.ylabel('Price in USD')
plt.show()

#Predict the price of a house with 4000 square feet
print('The price of a house with 4000 square feet can be:', round(lin_reg.predict([[4000]])[0][0],2), 'USD')
```

After doing the exercise, I can deduce that house prices in the United States increase linearly based on the size of their land as we can see in Figure 1. Also, we can visualize in Figure 2 that it doesn't matter much what type of neighborhood it is. In the 3 types of neighborhood the prices vary.
